Run this notebook to convert the Keras `.h5` model to a TensorFlow.js model for the web app.

```python
# 1. Install Dependencies
!pip install tensorflowjs tensorflow h5py
```

```python
# 2. Create Output Directory
import os
output_dir = "public/models/emotion_model"
os.makedirs(output_dir, exist_ok=True)
print(f"Created directory: {output_dir}")
```

```python
# 3. Run Conversion
# Ensure the input path is correct relative to where you run this notebook
input_path = "stress_ai/Stress-Detection-using-ML-and-Image-Processing-Techniques/model.h5"
output_path = "public/models/emotion_model"

!tensorflowjs_converter --input_format=keras "{input_path}" "{output_path}"
```

```python
# 4. Verify Output
if os.path.exists(f"{output_path}/model.json"):
    print("✅ Success! model.json created.")
    print("Files generated:", os.listdir(output_path))
else:
    print("❌ Error: model.json not found. Check the conversion output above for errors.")
```
